# Insights

The "Insights" section delves deeper into the dataset to extract meaningful insights and observations based on the previous analysis. It uses calculations, visualizations, and statistical techniques to uncover patterns, correlations, and potential explanations for observed trends.

The following code calculates the average time taken to ship orders.   It converts 'Order Date' and 'Ship Date' to datetime objects, calculates the difference in days, and then computes the average shipping time. This provides insight into the typical shipping duration

```python
# Convert 'Order Date' and 'Ship Date' to datetime objects
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Calculate the difference between Ship Date and Order Date
df['Shipping Time'] = (df['Ship Date'] - df['Order Date']).dt.days

# Calculate the average shipping time
average_shipping_time = df['Shipping Time'].mean()

# Display the result
print(f"Average Shipping Time (days): {average_shipping_time}")
```

The following code identifies the most profitable product category. It groups the data by 'Category', calculates the total profit for each category, and then determines the category with the highest profit. This helps understand which category contributes most to the company's profitability.

```python
# Calculate the profit for each category
category_profit = df.groupby('Category')['Profit'].sum()

# Display the result
print("Profit for each Category:")
print(category_profit)
most_profitable_category = category_profit.idxmax()

# Print the result
print(f"\nThe most profitable category is: {most_profitable_category}")
```

The following code assesses the association between shipping mode and region using a chi-squared test. It creates a contingency table to represent the frequency of each combination and then performs the test to determine if there is a statistically significant relationship. This helps understand if certain shipping modes are preferred in specific regions.

```python
# Create a contingency table
contingency_table = pd.crosstab(df['Ship Mode'], df['Region'])

# Perform chi-squared test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Contingency Table:")
print(contingency_table)
print("\nChi-squared statistic:", chi2)
print("P-value:", p)

# Interpret the results
alpha = 0.05  # Significance level
if p < alpha:
    print("\nThere is a statistically significant association between shipping mode and region.")
else:
    print("\nThere is no statistically significant association between shipping mode and region.")
```

The following code visualizes the regional distribution of specific sub-categories using heatmaps. It filters the data for each sub-category, creates contingency tables, and then generates heatmaps to show the frequency of sales in each region. This helps identify regional preferences or patterns for specific products.

```python
# Filter data for Bookcases sub-category
bookcases_df = df[df['Sub-Category'] == 'Bookcases']

# Create a contingency table
contingency_table = pd.crosstab(bookcases_df['Region'], bookcases_df['Sub-Category'])

# Create the heatmap
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
sns.heatmap(contingency_table, annot=True, cmap='viridis', fmt='d')  # fmt='d' for integers
plt.title('Distribution of Bookcases among Regions')
plt.xlabel('Sub-Category')
plt.ylabel('Region')
plt.show()
```

```python
# Filter data for Supplies sub-category
supplies_df = df[df['Sub-Category'] == 'Supplies']

# Create a contingency table
contingency_table = pd.crosstab(supplies_df['Region'], supplies_df['Sub-Category'])

# Create the heatmap
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
sns.heatmap(contingency_table, annot=True, cmap='viridis', fmt='d')  # fmt='d' for integers
plt.title('Distribution of Supplies among Regions')
plt.xlabel('Sub-Category')
plt.ylabel('Region')
plt.show()
```

```python
# Filter data for Tables sub-category
tables_df = df[df['Sub-Category'] == 'Tables']

# Create a contingency table
contingency_table = pd.crosstab(tables_df['Region'], tables_df['Sub-Category'])

# Create the heatmap
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
sns.heatmap(contingency_table, annot=True, cmap='viridis', fmt='d')  # fmt='d' for integers
plt.title('Distribution of Tables among Regions')
plt.xlabel('Sub-Category')
plt.ylabel('Region')
plt.show()
```
